```python
import numpy as np
import matplotlib.pyplot as plt
import superpose

# define the dictionary of sine wave parameters
sine_params = {
    'sine1': {'amplitude': 1, 'freq': 1, 'phase': 0},
    'sine2': {'amplitude': 0.5, 'freq': 2, 'phase': np.pi/2},
    'sine3': {'amplitude': 0.2, 'freq': 5, 'phase': np.pi},
}

# convert dictionary of dictionaries to list of dictionaries
sine_list = []
for key, params in sine_params.items():
    sine_dict = {'amplitude': params['amplitude'], 
                 'frequency': params['freq'], 
                 'phase': params['phase']}
    sine_list.append(sine_dict)

# generate and plot signal
start_idx = 2000
end_idx = 15000
sampling_rate = 10000
signal, t = superpose.superimpose_sine_waves(sine_list, sampling_rate, start_idx, end_idx)

plt.plot(t, signal)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Superimposed Sine Waves')
plt.show()
```
