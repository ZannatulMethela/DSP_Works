```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the sine waves
amplitudes = [1.0, 0.5, 0.2, 0.1]   
freq = 10.0                
phases = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # initial phases of the sine waves

# Create a time vector
t = np.linspace(0, 10, 10000)
print('t=',t)

y1 = amplitudes[0] * np.sin(2*np.pi*freq * t + phases[0])
y2 = amplitudes[1] * np.sin(2*np.pi*freq * t + phases[1])
y3 = amplitudes[2] * np.sin(2*np.pi*freq * t + phases[2])
y4 = amplitudes[3] * np.sin(2*np.pi*freq * t + phases[3])

plt.plot(t, y1)
plt.plot(t, y2)
plt.plot(t, y3)
plt.plot(t, y4)
plt.show()


# Generate the four sine waves
sine_waves = []
for i in range(4):
    sine_wave = amplitudes[i] * np.sin(2*np.pi*freq * t + phases[i])
    sine_waves.append(sine_wave)
#print('sine wave = ', sine_wave)
# Combine the sine waves into a superposition
superposition = sum(sine_waves)

#print('superposition', superposition)

# Plot the individual sine waves and the superposition
plt.figure()
plt.plot(t, sine_waves[0], label='Amp = 1.0, Phase = 0')
plt.plot(t, sine_waves[1], label='Amp = 0.5, Phase = pi/4')
plt.plot(t, sine_waves[2], label='Amp = 0.2, Phase = pi/2')
plt.plot(t, sine_waves[3], label='Amp = 0.1, Phase = 3pi/4')
plt.plot(t, superposition, label='Superposition')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
