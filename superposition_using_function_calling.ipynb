```python
import numpy as np
import matplotlib.pyplot as plt

def superimpose_sine_waves(sine_list, sampling_rate, start_idx, end_idx):
    t = np.arange(0, 1, 1/sampling_rate) # generate time values
    signal = np.zeros_like(t) # initialize signal to 0
    for sine_dict in sine_list:
        amplitude = sine_dict['amplitude']
        frequency = sine_dict['frequency']
        phase = sine_dict['phase']
        sine_wave = amplitude * np.sin(2*np.pi*frequency*t + phase)
        signal += sine_wave # add sine wave to signal
    return signal[start_idx:end_idx], t[start_idx:end_idx]

# define the dictionary of sine wave parameters
sine_params = {
    'sine1': {'amplitude': 1, 'freq': 1, 'phase': 0},
    'sine2': {'amplitude': 0.5, 'freq': 2, 'phase': np.pi/2},
    'sine3': {'amplitude': 0.2, 'freq': 5, 'phase': np.pi},
}

# convert dictionary of dictionaries to list of dictionaries
sine_list = []
for key, params in sine_params.items():
    sine_dict = {'amplitude': params['amplitude'], 
                 'frequency': params['freq'], 
                 'phase': params['phase']}
    sine_list.append(sine_dict)

# generate and plot signal
start_idx = 2000
end_idx = 15000
signal, t = superimpose_sine_waves(sine_list, 10000, start_idx, end_idx)

plt.plot(t, signal)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Superimposed Sine Waves')
plt.show()
```
